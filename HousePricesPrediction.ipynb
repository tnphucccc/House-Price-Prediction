# Data Processing

```python
data_na = data.dropna(subset=["total_bedrooms"])
data_na.shape
```

```python
dummies = pd.get_dummies(data_na.ocean_proximity)
data_na_dummies = pd.concat([data_na, dummies], axis=1)
data_finalized = data_na_dummies.drop(['ocean_proximity'], axis=1)
data_finalized
```

Split the Dataset

```python
from sklearn.model_selection import train_test_split

x = data_finalized.drop(['median_house_value'], axis=1)
y = data_finalized['median_house_value']
x.shape, y.shape
```

```python
"""
Explain the StandardScaler
standardize features by removing the mean and scaling to unit variance
The standard score of a sample x is calculated as:
z = (x - u) / s
where u is the mean of the training samples or zero if with_mean=False, and s is the standard deviation of the training samples or one if with_std=False.
"""
from sklearn.preprocessing import StandardScaler

"""
In this case, the fit method is calculating the mean and variance of each of the features present in our data. The transform method is transforming all the features using the respective mean and variance.
"""
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

# Building ML Models

```python
from sklearn.metrics import root_mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
import time
import tensorflow as tf
```

## Linear Regression

```python
model_linear = LinearRegression()

start_train = time.time()
model_linear.fit(x_train, y_train)
end_train = time.time()
train_time = end_train - start_train

start_test = time.time()
y_pred = model_linear.predict(x_test)
end_test = time.time()
test_time = end_test - start_test

print(f"Training time: {train_time:.4f} seconds --- Testing time: {test_time:.4f} seconds")
print('---' * 22)

root_mean_squared_error(y_test, y_pred), root_mean_squared_error(y_train, model_linear.predict(x_train))
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(x=y_test.index, y=y_test, label='Actual')
sns.lineplot(x=y_test.index, y=y_pred.flatten(), label='Predicted')
plt.legend()
plt.show()
```

## K Nearest Neighbor

```python
model_knn = KNeighborsRegressor(n_neighbors=10)

start_train = time.time()
model_knn.fit(x_train, y_train)
end_train = time.time()
train_time = end_train - start_train

start_test = time.time()
y_pred = model_knn.predict(x_test)
end_test = time.time()
test_time = end_test - start_test

print(f"Training time: {train_time:.4f} seconds --- Testing time: {test_time:.4f} seconds")
print('---' * 22)

root_mean_squared_error(y_test, y_pred), root_mean_squared_error(y_train, model_knn.predict(x_train))
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(x=y_test.index, y=y_test, label='Actual')
sns.lineplot(x=y_test.index, y=y_pred.flatten(), label='Predicted')
plt.legend()
plt.show()
```

## Random Forest Regressor

```python
# Reduce overfitting of the model
model_rf = RandomForestRegressor(n_estimators=200, max_depth=10, )

start_train = time.time()
model_rf.fit(x_train, y_train)
end_train = time.time()
train_time = end_train - start_train

start_test = time.time()
y_pred = model_rf.predict(x_test)
end_test = time.time()
test_time = end_test - start_test

print(f"Training time: {train_time:.4f} seconds --- Testing time: {test_time:.4f} seconds")
print('---' * 22)

root_mean_squared_error(y_test, y_pred), root_mean_squared_error(y_train, model_rf.predict(x_train))
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(x=y_test.index, y=y_test, label='Actual')
sns.lineplot(x=y_test.index, y=y_pred.flatten(), label='Predicted')
plt.legend()
plt.show()
```

## Neural Network
Explaining the model:
- Input layer has 13 features
- Hidden layer 1 has 256 neurons with ReLU activation function
- Hidden layer 2 has 128 neurons with ReLU activation function
- Hidden layer 3 has 64 neurons with ReLU activation function
- Hidden layer 4 has 32 neurons with ReLU activation function
- Output layer has 1 neuron with linear activation function

Optimizer: Adam

Loss function: Mean Squared Error

Metrics: Root Mean Squared Error

Epochs: 100

Batch size: 32

Validation data: x_test, y_test

```python
model_nn = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(13,)),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='linear')
])

model_nn.compile(optimizer='adam', loss='mean_squared_error', metrics=['root_mean_squared_error'])

start_train = time.time()
model_nn.fit(x_train, y_train, epochs=100, batch_size=32, validation_data=(x_test, y_test))
end_train = time.time()
train_time = end_train - start_train

start_test = time.time()
y_pred = model_nn.predict(x_test)
end_test = time.time()
test_time = end_test - start_test

print(f"Training time: {train_time:.4f} seconds --- Testing time: {test_time:.4f} seconds")
print('---' * 22)

root_mean_squared_error(y_test, y_pred), root_mean_squared_error(y_train, model_nn.predict(x_train))
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(x=y_test.index, y=y_test, label='Actual')
sns.lineplot(x=y_test.index, y=y_pred.flatten(), label='Predicted')
plt.legend()
plt.show()
```

## Evaluation

Plot the Root Mean Squared Error for each model
to evaluate the performance between them.
The lower the value, the better the model

```python
import numpy as np

models = ['Linear Regression', 'KNN', 'Random Forest', 'Neural Network']
test_rmse = [
    root_mean_squared_error(y_test, model_linear.predict(x_test)),
    root_mean_squared_error(y_test, model_knn.predict(x_test)),
    root_mean_squared_error(y_test, model_rf.predict(x_test)),
    root_mean_squared_error(y_test, model_nn.predict(x_test))
]
train_rmse = [
    root_mean_squared_error(y_train, model_linear.predict(x_train)),
    root_mean_squared_error(y_train, model_knn.predict(x_train)),
    root_mean_squared_error(y_train, model_rf.predict(x_train)),
    root_mean_squared_error(y_train, model_nn.predict(x_train))
]

x = np.arange(len(models))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(16, 9))
rects1 = ax.bar(x - width/2, test_rmse, width, label='Test Data', color='skyblue')
rects2 = ax.bar(x + width/2, train_rmse, width, label='Train Data', color='midnightblue')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('RMSE')
ax.set_title('Root Mean Squared Error by model and data set')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

fig.tight_layout()

plt.show()
```

```python
# Use lineplot to plot the actual vs predicted values
plt.figure(figsize=(16, 9))
sns.lineplot(x=y_test.index, y=y_test, label='Actual')
sns.lineplot(x=y_test.index, y=model_linear.predict(x_test).flatten(), label='Linear Regression')
sns.lineplot(x=y_test.index, y=model_knn.predict(x_test), label='KNN')
sns.lineplot(x=y_test.index, y=model_rf.predict(x_test), label='Random Forest')
sns.lineplot(x=y_test.index, y=model_nn.predict(x_test).flatten(), label='Neural Network')
plt.title("Actual vs Predicted")
plt.show()
```
